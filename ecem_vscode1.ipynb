```python
filtered_data=data1[
    (data1['statut_dossier'].str.contains('Clos', case=False, na=False)) &
    (data1['statut_dossier'] == 'Réalisation totale')
]

grouped_data = filtered_data.groupby('Nom Officiel Département Majuscule')


total_closed_dossiers = data1.groupby('Nom Officiel Département Majuscule').size()
realisation_totale_dossiers = grouped_data.size()


percentage_realisation_totale = (realisation_totale_dossiers / total_closed_dossiers) * 100
percentage_realisation_totale
```

'statut_dossier' column contains the word 'Clos' and the 'statut_dossier' is exactly equal to 'Réalisation totale'.only there rows are included in the filtered data column and in the grouped data it select the filtered data based on different departments.total_closed_dossiers: This variable calculates the total number of closed dossiers for each department.realisation_totale_dossiers: This variable calculates the number of closed dossiers with 'Réalisation totale' status for each department. percentage_realisation_totale: This variable calculates the percentage of closed dossiers
